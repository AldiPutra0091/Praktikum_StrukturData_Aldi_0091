```python
def bubble_sort(Arr):
    n = len(Arr)

    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if Arr[j] > Arr[j + 1]:
                Arr[j], Arr[j + 1] = Arr[j + 1], Arr[j]

# Contoh penggunaan
V = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(V)
print("Hasil Bubble Sort dari data diatas adalah", V)
```

Bubble Sort adalah algoritma pengurutan sederhana yang secara berulang membandingkan dan menukar elemen-elemen adjacent jika mereka berada dalam urutan yang salah.

```python
def selection_sort(Arr):
    n = len(Arr)

    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if Arr[j] < Arr[min_index]:
                min_index = j

        Arr[i], Arr[min_index] = Arr[min_index], Arr[i]

# Contoh penggunaan
V = [64, 34, 25, 12, 22, 11, 90]
selection_sort(V)
print("Hasil Selection Sort dari data diatas adalah", V)
```

Selection Sort adalah algoritma pengurutan yang pada setiap iterasi mencari elemen minimum dari sisa array dan menukarnya dengan elemen pertama yang belum terurut.



```python
def insertion_sort(Arr):
    for i in range(1, len(Arr)):
        key = Arr[i]
        j = i - 1
        while j >= 0 and key < Arr[j]:
            Arr[j + 1] = Arr[j]
            j -= 1
        Arr[j + 1] = key

# Contoh penggunaan
V = [64, 34, 25, 12, 22, 11, 90]
insertion_sort(V)
print("Hasil Insertion Sort dari data diatas adalah", V)
```

Insertion Sort adalah algoritma pengurutan yang membagi array menjadi dua bagian: bagian terurut dan bagian tidak terurut. Pada setiap langkah, elemen dari bagian tidak terurut diambil dan dimasukkan ke dalam bagian terurut sesuai dengan urutan yang benar.
