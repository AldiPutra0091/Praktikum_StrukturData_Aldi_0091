```python
def is_pangram(sentence):
    # Inisialisasi set untuk menyimpan huruf-huruf yang sudah muncul
    alphabet_set = set()

    # Konversi kalimat ke huruf kecil untuk memastikan kesamaan huruf
    sentence = sentence.lower()

    # Iterasi melalui setiap karakter dalam kalimat
    for char in sentence:
        # Cek apakah karakter merupakan huruf dan belum ada dalam set
        if char.isalpha() and char not in alphabet_set:
            # Tambahkan karakter ke dalam set
            alphabet_set.add(char)

    # Cek apakah set huruf yang muncul memiliki jumlah huruf sama dengan total huruf dalam alfabet
    return len(alphabet_set) == 26

# Contoh penggunaan
kalimat = "Faqih dan Yazid meragukan objektivitas wacana profesor xilologi"
if is_pangram(kalimat):
    print(f'"{kalimat}" adalah pangram.')
else:
    print(f'"{kalimat}" bukan pangram.')
```
